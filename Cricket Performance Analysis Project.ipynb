```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")
matches.head(5)
deliveries.head(5)
```

```python
# 1. Data Cleaning
matches_cleaned = matches.drop(columns=['method', 'umpire1', 'umpire2'])
matches_cleaned['city'].fillna('Unknown', inplace=True)
matches_cleaned['player_of_match'].fillna('No Award', inplace=True)
matches_cleaned['winner'].fillna('No Result', inplace=True)
matches_cleaned['result_margin'].fillna(0, inplace=True)
matches_cleaned.head(5)
```

```python
deliveries_cleaned = deliveries.drop(columns=['extras_type', 'fielder'])
deliveries_cleaned['player_dismissed'].fillna('Not Out', inplace=True)
deliveries_cleaned['dismissal_kind'].fillna('None', inplace=True)

deliveries_cleaned.head(5)
```

```python
# Standardizing Team Names
team_name_mapping = {
    "Royal Challengers Bangalore": "Royal Challengers Bengaluru",
    "Delhi Daredevils": "Delhi Capitals",
    "Deccan Chargers": "Sunrisers Hyderabad" ,
    "Rising Pune Supergiants": "Rising Pune Supergiant"  
}

# Replacing team names in all relevant columns
matches_cleaned['team1'] = matches_cleaned['team1'].replace(team_name_mapping)
matches_cleaned['team2'] = matches_cleaned['team2'].replace(team_name_mapping)
matches_cleaned['winner'] = matches_cleaned['winner'].replace(team_name_mapping)

# Exclude 'No Result' from winners
team_wins = matches_cleaned[matches_cleaned['winner'] != 'No Result']['winner'].value_counts()

plt.figure(figsize=(12,6))
ax = sns.barplot(y=team_wins.index, x=team_wins.values, palette="viridis")
plt.xlabel("Total Wins")
plt.ylabel("Teams")
plt.title("Most Successful IPL Teams")

# Adding value labels on bars
for index, value in enumerate(team_wins.values):
    ax.text(value + 2, index, str(value), va='center', fontsize=12)

plt.show()
```

```python
# 3. Top Batsmen - Total Runs
# Calculate total runs for each batsman
batsman_runs = deliveries_cleaned.groupby('batter')['batsman_runs'].sum().reset_index()

# Get top 10 batsmen by total runs
top_batsmen = batsman_runs.sort_values(by='batsman_runs', ascending=False).head(10)

# Plot
plt.figure(figsize=(12,6))
ax = sns.barplot(y=top_batsmen['batter'], x=top_batsmen['batsman_runs'], palette="coolwarm")
plt.xlabel("Total Runs")
plt.ylabel("Batsmen")
plt.title("Top 10 IPL Batsmen by Runs")

# Adding labels on bars
for index, value in enumerate(top_batsmen['batsman_runs']):
    ax.text(value + 50, index, str(value), va='center', fontsize=12)

plt.show()

```

```python
# 4. Top Bowlers - Total Wickets
wicket_deliveries = deliveries_cleaned[deliveries_cleaned['player_dismissed'] != 'Not Out']
bowler_wickets = wicket_deliveries.groupby('bowler')['player_dismissed'].count().reset_index()
bowler_wickets.rename(columns={'player_dismissed': 'wickets'}, inplace=True)
top_bowlers = bowler_wickets.sort_values(by='wickets', ascending=False).head(10)
plt.figure(figsize=(12,6))
ax = sns.barplot(y=top_bowlers['bowler'], x=top_bowlers['wickets'], palette="magma")
plt.xlabel("Total Wickets")
plt.ylabel("Bowlers")
plt.title("Top 10 IPL Bowlers by Wickets")

# Adding labels on bars
for index, value in enumerate(top_bowlers['wickets']):
    ax.text(value + 1, index, str(value), va='center', fontsize=12)

plt.show()
```

```python
# Convert venue names to lowercase and remove leading/trailing spaces
matches['venue'] = matches['venue'].str.lower().str.strip()

# Fix known duplicate venue names
venue_name_mapping = {
    "wankhede stadium, mumbai": "wankhede stadium",
    "m. a. chidambaram stadium, chepauk": "m. a. chidambaram stadium",
    "maharashtra cricket association stadium": "mca stadium"
}
matches['venue'] = matches['venue'].replace(venue_name_mapping)

# Count the most common venues
venue_counts = matches['venue'].value_counts().head(10)

# Plot
plt.figure(figsize=(12,6))
ax = sns.barplot(y=venue_counts.index, x=venue_counts.values, palette="viridis")
plt.xlabel("Number of Matches")
plt.ylabel("Venue")
plt.title("Top 10 Most Common IPL Venues ")

# Adding labels on bars
for index, value in enumerate(venue_counts.values):
    ax.text(value + 2, index, str(value), va='center', fontsize=12)

plt.show()
```

```python
# Check if 'match_type' column exists, otherwise identify final matches manually
if 'match_type' in matches.columns:
    final_matches = matches[matches['match_type'] == 'Final']
else:
    final_matches = matches.sort_values(by=['season', 'match_id']).groupby('season').last().reset_index()

# Count final appearances for each team, fill NaN with 0
final_appearances = final_matches['team1'].value_counts() + final_matches['team2'].value_counts()
final_appearances = final_appearances.fillna(0).astype(int)  # Replace NaN with 0 and convert to int

# Sort values and select top 5
final_appearances = final_appearances.sort_values(ascending=False).head(5)

# Visualization
plt.figure(figsize=(12,6))
ax = sns.barplot(y=final_appearances.index, x=final_appearances.values, palette="coolwarm")

plt.xlabel("Number of Final Appearances")
plt.ylabel("Teams")
plt.title("Top 5 Teams with Most IPL Final Appearances")

# Adding labels on bars (ensuring integer format)
for index, value in enumerate(final_appearances.values):
    ax.text(value + 0.2, index, str(value), va='center', fontsize=12)  # Ensuring labels as int

plt.show()
```

```python

# Count top 10 players with most 'Player of the Match' awards
player_awards = matches['player_of_match'].value_counts().head(10)

# Visualization
plt.figure(figsize=(12,6))
ax = sns.barplot(y=player_awards.index, x=player_awards.values, palette="magma")

plt.xlabel("Total Awards")
plt.ylabel("Players")
plt.title("Top 10 Players with Most Player of the Match Awards")

# Adding labels on bars (ensuring integer format)
for index, value in enumerate(player_awards.values):
    ax.text(value + 0.2, index, str(int(value)), va='center', fontsize=12)  # Convert to int

plt.show()
```

```python
#Highest and Lowest Total Scores in IPL
# Aggregate total runs per match per team
match_scores = deliveries.groupby(['match_id', 'batting_team'])['total_runs'].sum().reset_index()

# Get the top 10 highest and lowest scores
highest_scores = match_scores.sort_values(by='total_runs', ascending=False).head(10)

# Create a unique label for each match (Team + Score)
highest_scores["label"] = highest_scores["batting_team"] + " (" + highest_scores["total_runs"].astype(str) + ")"

plt.figure(figsize=(12,6))
sns.barplot(y=highest_scores['label'], x=highest_scores['total_runs'], palette="Greens_r")
plt.xlabel("Total Runs")
plt.ylabel("Teams")
plt.title("Top 10 Highest Team Scores in IPL History")
plt.show()


```

```python
# Most Economical Bowlers (Min 300 Overs)
bowler_economy = deliveries.groupby('bowler').agg({'total_runs': 'sum', 'over': 'count'})
bowler_economy['economy'] = bowler_economy['total_runs'] / (bowler_economy['over'] / 6)

economical_bowlers = bowler_economy[bowler_economy['over'] > 300].sort_values(by='economy').head(10)

plt.figure(figsize=(12,6))
ax = sns.barplot(y=economical_bowlers.index, x=economical_bowlers['economy'], palette="Blues_r")
plt.xlabel("Economy Rate")
plt.ylabel("Bowlers")
plt.title("Top 10 Most Economical Bowlers (Min 300 Overs)")

# Adding value labels
for index, value in enumerate(economical_bowlers['economy']):
    ax.text(value + 0.1, index, f"{value:.2f}", va='center', fontsize=12)

plt.show()
```
